```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Testing different libraries for map handling
import geopandas as gpd
```

```python
places = pd.read_csv('places.csv')
#Getting data for plotting world map
countries = gpd.read_file('./natural_earth/ne_110m_admin_0_countries.shx')
```

```python
plt.figure()
countries.plot(
    edgecolor = 'lightgray', cmap = 'viridis',
    figsize = (15,10), column = 'CONTINENT').set_title(
    'Touristing!', fontweight = 'bold', fontsize = 18)
sns.scatterplot(places, x = 'lon', y = 'lat', legend = False,
                hue = 'country', palette = 'bright')
plt.grid()
```
